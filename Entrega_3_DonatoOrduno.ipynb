### Ejercicio 4

Imagina que tienes tres cuerpos en un plano bidimensional.  Las fuerzas que actuan sobre cada cuerpo se pueden representar en una matriz.  Cada fila de la matriz representa un cuerpo y cada columna representa una fuerza en una dirección concreta ( per ejemplo, fuerza en la dirección X y fuerza en la dirección Y) en Newtons.

Las fuerzas que actuan sobre los cuerpos son las seguientes:
- Cuerpo A: Fuerza X = 10 y Fuerza Y = 5.
- Cuerpo B: Fuerza X = 4 y Fuerza Y = 12.
- Cuerpo C: Fuerza X = -6 y Fuerza Y = -8.

Responde a las seguientes preguntas:
- a) Crea la matriz de fuerzas y muéstrala por pantalla. **(0.25 puntos)**
- b) Calcula la fuerza resultante sobre cada cuerpo. La fuerza resultante se puede calcular como la suma vectorial de las fuerzas.  Para hacerlo, utiliza una función de `numpy`. **(0.5 puntos)**
- c) Calcula la aceleración de cada cuerpo mediante la segunda ley de Newton, `F=m*a`, considerando que la masa de cada cuerpo es de 5 kg. **(0.25 puntos)**

**Nota**.  Los resultados deben presentarse con sólo dos decimales.

**(Total 1 punto)** $\color{Orange}{\text{EG}}$

```python
# Apartado a) creamos la matriz de fuerzas
import numpy as np
force_matrix = np.matrix(([10,5],[4,12],[-6,-8]))
print(force_matrix.shape)
print(type(force_matrix))
print(force_matrix)
```

```python
# Apartado b) Calculamos la fuerza resultante
## Magnitud vectorial
force_results = []

for x in force_matrix:
   force_results.append(float(np.linalg.norm(x)))

print(force_results)
```

```python
# Apartado c)
## F=m*a   m = 5kg
force_results = np.array(force_results)
a = force_results / 5
print(a)
```

```python
# Respuesta C: Salario medio mensual por género y departamento (barras agrupadas)

dep_salaries_gender_2 = employees.pivot_table(values=['Salary','AnnualSalary'], index='Department', columns='Gender')
print(dep_salaries_gender_2)

(dep_salaries_gender_2.reset_index(inplace=True))
dep_salaries_gender_2

fig , ax = plt.subplots(nrows=2, figsize = [6,9])


dep_salaries_gender_2.plot(kind='bar',
                           x = 'Department',
                           y = 'Salary',
                           colormap = 'Set3',
                           ylim = [3000,5000],
                           xlabel = 'Department',
                           ylabel = 'Average salary (€)',
                           ax = ax[0])
ax[0].set_title('Average salary by gender and department')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=25, ha='right')
ax[0].legend(loc = 'upper left')




dep_salaries_gender_2.plot(kind='bar',
                           x = 'Department',
                           y = 'AnnualSalary',
                           colormap = 'Set3',
                           ylim = [35000, 60000],
                           xlabel = 'Department',
                           ylabel = 'Average salary (€)',
                           ax = ax[1])
ax[1].set_title('Average annual salary by gender and department')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=25, ha='right')
ax[1].legend(loc = 'upper left')

plt.tight_layout(h_pad=2)
plt.show()
```

```python
# Respuesta D: Distribución de rendimiento (histograma)
employees.hist(column='PerformanceScore',
               bins = 10,
               rwidth = 0.9,)
plt.xlabel('Performance score')
plt.ylabel('Frequency')
plt.title('Distribution of staff performance')
plt.show()
```

- **b)** Realiza un **test de normalidad** utilizando la función [`shapiro`](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html) de `scipy.stats`, con el objetivo de comprobar si los datos siguen una distribución normal.  
  Muestra por pantalla el **p-value** e interpreta brevemente el resultado obtenido, indicando si puede asumirse o no la normalidad de los datos.  **(1 punto)**  $\color{orange}{\text{EG}}$  

- **c)** Representa un **histograma** de los datos generados para visualizar su distribución. **(0.5 puntos)**  $\color{green}{\text{NM}}$  

- **d)** De manera **opcional**, utiliza la librería **`seaborn`** para mostrar la **curva de densidad de probabilidad** (`kdeplot`) y comparar visualmente si los datos parecen seguir una distribución normal.  **Opcional.**  $\color{red}{\text{EI}}$  

**(Total: 2 puntos)**

```python
# Respuesta A
from scipy.stats import norm
datos = norm.rvs(loc = 50, scale = 10, size = 100)
```

```python
# Respuesta B: Test de normalidad
from scipy.stats import shapiro
shapiro(datos)[1]
```

Si se puede asumir normalidad ya que el p-value es mayor al valor de significancia (por lo general 0.05).

```python
# Respuesta C:  Histograma de la distribución
plt.hist(datos,
         rwidth= 0.9)
plt.xlabel('values')
plt.ylabel('percentage of distribution')
plt.show()
```

```python
# Respuesta D: (Opcional) Visualización con seaborn
import seaborn as sns

sns.kdeplot(datos)
plt.xlabel('values')
plt.title('Probability density curve')
plt.show()
```
